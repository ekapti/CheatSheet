## If function

```
def calc(x, y, op="None"):         # three parameters (without any defaults)
    if op == 'add':         # conditional statement
        return x + y
    elif op == 'subtract':
        return x - y
    elif op == 'multiply':
        return x * y
    else:
        print ('Valid operations: add, subtract, multiply')
```

## How to write a for loop and append?

```
# for loop to create a list of 2, 4, 6, 8, 10
nums= [1,2,3,4,5]
doubled = []                # create empty list to store results
for num in nums:            # loop through nums (will execute 5 times)
    doubled.append(num*2)   # append the double of the current value of num

doubled
```

```
# or (also called list comprehension)
doubled = [num*2 for num in nums ]
doubled
```

#### Function to find max of a list using reduce.

```
def max_of(a,b):
  if (a>b):
    return(a)
  else:
    return(b)
```

```
list= [1,2,3,4,5,6]
max(list)
```

```
max_list = reduce(max_of, list, 1)
max_list
```

##### Important: you can not use reduce to calculate sum of squares since it will not add square of the element, so you need to square it and place it into the top of head. 

###### With lambda

```
ll = [2,3,4]
product_list = map(lambda x: x**2, ll)
product_list
```

```
product_sum = sum(map(lambda x : x**2,ll ))
product_sum
```

###### With a for loop

```
product_sum = 0
for num in ll:
  product_sum = product_sum + num**2
product_sum
```

###### With reduce function


```
# it is not taking the square of the first value, 
#for that reason when you operate on a value you should not use reduce.
product_sum = reduce(lambda acc, val: acc+val**2, ll)
product_sum
```

```
# you need to create a list by squaring the first value
product_sum = reduce(lambda acc, val: acc+val**2, [ll[:1][0]**2]+ll[1:])
product_sum
```
